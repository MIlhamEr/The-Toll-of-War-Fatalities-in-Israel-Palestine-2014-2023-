## Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

## Dataset

GA PERLU UPLOAD LAGI DATASETNYA YA, ini tinggal di Run aja (dah yang versi cleaned_dataset)

```python
url = 'https://raw.githubusercontent.com/JCalvinL19/Dataset/main/Cleaned_Fatalities_ISR_PSE_Conflict_2000_to_2023.csv'
df = pd.read_csv(url)
df.head()
```

```python
df['date_of_death'] = pd.to_datetime(df['date_of_death'])
df['year_of_death'] = df['date_of_death'].dt.year
df['year_of_death']
```

```python
df = df[(df['year_of_death'] >= 2014)]
df.head()
```

```python
df.info()
```

```python
df.shape
```

## Visualisasi Data dengan Python

### Usia & Gender Korban

```python
sns.histplot(df['age'], bins=20, kde=True)
plt.title('Histogram Usia')
plt.show()
```

```python
gender_counts = df['gender'].value_counts()
gender_counts
```

```python
sns.barplot(x=gender_counts.index, y=gender_counts.values)
plt.title('Jumlah Individu Berdasarkan Jenis Kelamin')
plt.show()
```

```python
sns.violinplot(x='gender', y='age', data=df)
plt.title('Sebaran Usia Berdasarkan Jenis Kelamin')
plt.show()
```

```python
fig = plt.figure(figsize=(12, 5))
sns.countplot(x='age_category', hue='gender', data=df)
plt.title('Jumlah Individu Berdasarkan Kategori Usia dan Jenis Kelamin')
plt.show()
```

### Kewarganegaraan Korban

```python
citizenship_counts = df['citizenship'].value_counts()
citizenship_counts
```

```python
exclude_conditions = ~df['citizenship'].isin(['American', 'Jordanian'])
citizen_data = df[exclude_conditions]
```

```python
citizenship_counts = citizen_data['citizenship'].value_counts()
citizenship_counts
```

```python
plt.bar(citizenship_counts.index, citizenship_counts.values)
plt.title('Kewarganegaraan Korban Tewas')
plt.show()
```

### Kewarganegaraan Pelaku

```python
killed_by_counts = citizen_data['killed_by'].value_counts()
killed_by_counts
```

```python
plt.bar(killed_by_counts.index, killed_by_counts.values)
plt.title('Kewarganegaraan Yang Membunuh')
plt.show()
```

### Hubungan Kewarganegaraan Korban & Pelaku

```python
citizen_killed_data = citizen_data.groupby(['citizenship', 'killed_by']).size().reset_index(name='count')
citizen_killed_data
```

```python
sns.barplot(x='citizenship', y='count', hue='killed_by', data=citizen_killed_data)
plt.legend(title='Penyebab Kematian', loc=2)
plt.show()
```

### Lokasi

```python
locations_counts = citizen_data['event_location_district'].value_counts()
locations_counts
```

```python
plt.barh(locations_counts.index, locations_counts.values)
plt.show()
```

### Angka Kematian per Tahun

```python
years_counts = df['year_of_death'].value_counts().sort_index()
years_counts
```

```python
plt.figure(figsize=(10, 4))
plt.plot(years_counts.index, years_counts.values, marker='o', linestyle='-')
plt.title('Frekuensi Kematian per Tahun')
plt.grid(axis='y')
plt.show()
```

### Senjata & Jenis Cedera

```python
ammunition_counts = df['ammunition'].value_counts()
ammunition_counts
```

```python
plt.barh(ammunition_counts.index, ammunition_counts.values)
plt.show()
```

```python
injury_counts = df['type_of_injury'].value_counts()
injury_counts
```

```python
plt.barh(injury_counts.index, injury_counts.values)
plt.show()
```

```python
injury_data = citizen_data.groupby(['type_of_injury', 'ammunition']).size()
injury_data = injury_data.reset_index(name='count')
injury_data = injury_data.sort_values(by='count', ascending=False)
injury_data
```

```python
injury_data = injury_data[injury_data['count'] >= 10]
injury_data
```

```python
plt.figure(figsize=(10, 4))
sns.barplot(x='ammunition', y='count', hue='type_of_injury', data=injury_data)
plt.title('Perbandingan Jenis Cedera Berdasarkan Senjata')
plt.show()
```

```python
bomb_data = injury_data[injury_data['ammunition'] == "Bomb"]
bomb_data
```

```python
plt.figure(figsize=(6, 4))
sns.barplot(x='type_of_injury', y='count', data=bomb_data)
plt.title('Perbandingan Jenis Cedera Yang Disebabkan Bomb')
plt.show()
```

```python
top5_injury_data = injury_data[injury_data['count'] >= 50]
top5_injury_data
```

```python
plt.figure(figsize=(8, 4))
sns.barplot(x='ammunition', y='count', data=top5_injury_data)
plt.title('5 Senjata Teratas yang Digunakan')
plt.show()
```

### Keikutsertaan Dalam Perselisihan

```python
hostilities_counts = df['took_part_in_the_hostilities'].value_counts()
hostilities_counts
```

```python
plt.figure(figsize=(8, 4))
sns.barplot(x=hostilities_counts.index, y=hostilities_counts.values)
plt.title('Perbandingan Korban yang Terlibat Secara Langsung Dalam Perselisihan')
plt.show()
```

```python
hostilities_data = df.groupby(['took_part_in_the_hostilities', 'citizenship']).size().reset_index(name='count')
hostilities_data
```

```python
hostilities_data = hostilities_data[hostilities_data['count'] > 1]
hostilities_data
```

```python
plt.figure(figsize=(8, 4))
sns.barplot(x=hostilities_data['took_part_in_the_hostilities'],
            y=hostilities_data['count'],
            hue=hostilities_data['citizenship'])
plt.show()
```
